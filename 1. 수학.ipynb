```python
import pandas as pd
import matplotlib.pyplot as plt # 그래프 그리기
from matplotlib import rc # 한글폰트 사용
import matplotlib.font_manager as fm # 폰트 가져오기
import numpy as np

font_name = fm.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
```

```python
# 이차 함수

x = [i for i in range(-7, 14)]
y = [(i - 3) ** 2 + 1 for i in x]

plt.figure(figsize=(12,8))
plt.plot(x, y)
```

```python
def func(a, x):
    return a ** x

x = [i for i in range(1, 11)]
y = [func(2, i) for i in x]

plt.plot(x, y)
```

```python
import math

def logFunc(x):
    # math.log10()
    # math.log() => 밑이 e인 자연로그
    
    # 밑이 2인 Log
    return math.log2(x)

x = [i for i in range(1, 11)]
y = [logFunc(i) for i in x]

plt.plot(x,y)
```

```python
# 시그모이드 함수
# 자연 상수 e가 분포에 포함되어 있는 지수 함수
# f(x) = 1 / (1 + e ** -x)

import math
# math.e => 자연 상수

def func(x):
    return 1 / (1 + math.e ** -x)

x = [i for i in range(-10, 11)]
y = [func(i) for i in x]

plt.plot(x,y)
```

```python
# 미분

x = [i for i in range(1, 21)]
y = [i**2 for i in x]

plt.figure(figsize=(12,8))
plt.plot(x,y)
plt.plot([x[5],x[10]], [y[5],y[10]])

plt.scatter(x[5], y[5], s=100, color='black')
plt.annotate('A', (x[5], y[5]), fontsize=25)

plt.scatter(x[10], y[10], s=100, color='black')
plt.annotate('B', (x[10], y[10]), fontsize=25)

plt.show()
```
