```python
# 시그모이드 함수
# 자연 상수 e가 분포에 포함되어 있는 지수 함수
# f(x) = 1 / (1 + e ** -x)

import math
# math.e => 자연 상수

def func(x):
    return 1 / (1 + math.e ** -x)

x = [i for i in range(-10, 11)]
y = [func(i) for i in x]

plt.plot(x,y)
```

```python
import math

def func(a, b, x):
    return 1 / (1 + math.e ** -(a * x + b))

x = [i for i in range(-10, 11)]

y1 = [func(1, 0, i) for i in x]
y2 = [func(5, 0, i) for i in x]

plt.plot(x,y1, color='r', label='a = 1')
plt.plot(x,y2, color='b', label='a = 5')
plt.legend()

plt.show()


y3 = [func(1, -5, i) for i in x]
y4 = [func(1, 0, i) for i in x]
y5 = [func(1, 5, i) for i in x]

plt.plot(x,y3, color='r', label='b = -5')
plt.plot(x,y4, color='b', label='b = 0')
plt.plot(x,y5, color='g', label='b = 5')
plt.legend()

plt.show()
```

```python
import math

def func(a, b, x):
    return 1 / (1 + math.e ** -(a * x + b))

x = [i for i in range(-10, 11)]

y1 = [func(1, -5, i) for i in x]
y2 = [func(1, 0, i) for i in x]
y3 = [func(1, 5, i) for i in x]

plt.plot(x,y1, color='r', label='b = -5')
plt.plot(x,y2, color='g', label='b = 0')
plt.plot(x,y3, color='b', label='b = 5')
plt.legend()

plt.show()
```

```python
import math

x = [i for i in range(1, 101)]

y1 = [math.log2(i) for i in x]
y2 = [-math.log2(i) for i in x]

y3 = [math.log2(101 - i) for i in x]
y4 = [-math.log2(101 - i) for i in x]

plt.plot(x, y1, color='y', label = 'y = log2(i)')
plt.plot(x, y2, color='b', label = 'y = -log2(i)')
plt.plot(x, y3, color='g', label = 'y = log2(101 - i)')
plt.plot(x, y4, color='r', label = 'y = -log2(101 - 1)')

plt.legend()

plt.show()
```

```python
def sigmoid(f):
    return 1 / (1 + np.e ** (-f))


data = [
    [2, 0],
    [4, 0],
    [6, 0],
    [8, 1],
    [10, 1],
    [12, 1],
    [14, 1]
] # 0 : 불합격, 1 : 합격

x_data = [i[0] for i in data]
y_data = [i[1] for i in data]

plt.scatter(x_data, y_data, color = 'r')
plt.xlim(0, 15)
plt.ylim(-.1,1,1)

a = 0
b = 0

lr = 0.05


# 경사 하강법
for i in range(2001):
    for x, y in data:
        # ylog(h) + (1-y)log(1-h)를 a와 b에 대해 편미분한 공식
        a_diff = x * (sigmoid(a * x + b) - y)
        b_diff = sigmoid(a * x + b) - y
        
        a = a - lr * a_diff
        b = b - lr * b_diff
        
        if i % 1000 == 0:
            print(f'epoch = {i}, a = {a}, b = {b}')
    
x_range = (np.arange(0, 15, 0.1)) # 0 ~ 15, 0.1 단위
plt.plot(x_range, np.array([sigmoid(a * x + b) for x in x_range]))
```
