```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 각 정보간 어떤 상관관계가 있는지 분석을 도와주는 라이브러리
```

```python
# 타이타닉에 탑승한 사람들의 신상정보를 활용하여, 승선한 사람들의 생존 여부 예측

data = sns.load_dataset('titanic')
data
```

```python
# corr() 함수를 통해 각 항목이 얼마나 관련있는지 상관계수를 구할 수 있다.
data.corr()
```

```python
res = data.corr()

sns.clustermap(res,
               annot = True, # 화면 표시 여부
               cmap = 'RdYlBu_r', # Red, Yellow, Blue 색상으로 표시
               vmin = -1, vmax = 1 # -1 ~ 1의 범위의 밝기로 컬러 표시
              )

plt.show()
```
