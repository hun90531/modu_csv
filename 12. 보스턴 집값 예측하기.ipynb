```python
import pandas as pd
import numpy as np

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

from sklearn.model_selection import train_test_split
```

```python
# delim_whitespace = True는 공백으로 구분된 파일을 읽을 때 사용한다.

df_pre = pd.read_csv('deeplearning/dataset/housing.csv', delim_whitespace=True, header=None)
df_pre.head()
```

```python
df = df_pre.sample(frac=1) # 100%의 데이터를 랜덤으로 뽑기
df.head()
```

```python
dataset = df.values
x = dataset[:, :13]
y = dataset[:, 13]

print(x)
print('-' * 50)
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
model = Sequential()
model.add(Dense(30, input_dim = 13, activation = 'relu'))
model.add(Dense(6, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))

model.compile(loss = 'binary_crossentropy',
             optimizer='adam',
             metrics = ['accuracy']
             )
```

```python
model.fit(x_train, y_train, epochs=50, batch_size=10, verbose = False)
model.fit(x_train, y_train, epochs=3, batch_size=1, verbose = True)
```

```python
# 모델 예측

y_pre = model.predict(x_test).flatten()
print(list(y_pre[:10]))
print(list(y_test))
```

```python
result = []
for i in range(10):
    data = '실제가격 : %.3f\t예상가격:%.3f' % (y_test[i], y_pre[i])
    result.append(data)
    print(data)

print(result[:3])
```

```python
# 결과를 파일로 저장하기

with open('files/result_boston.txt', 'w') as f:
    for i in result:
        f.write(i + '\n')
```
