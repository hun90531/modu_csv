```python
# 최소 제곱법 2 

def top(x, mx,  y, my):
    d = 0
    for i in range(len(x)):
        # (x - x 평균)(y - y평 균)의 합
        d = d + (x[i] - mx) * (y[i] - my)
    
    return d

x = [2, 4, 6, 8]
y = [81, 93, 91, 97]

# 리스트 평균
mx = np.mean(x) # 5
my = np.mean(y) # 90.5

# (x - x평균)^2 의 합
divisor = sum([(mx - i) ** 2 for i in x]) # 20

dividend = top(x, mx, y, my) # 46

a = dividend / divisor # 2.3

# y 평균 - (x 평균 * a)
b = my - (mx * a) # 79

res = [a * i + b for i in x] # [83.6, 88.2, 92.8, 97.4]

plt.figure(figsize=(8,4))
plt.scatter(x, y, s=100)
plt.plot(x,res)
plt.show()
```

```python
def get_a(x, y, avg_x, avg_y):
    n, m = 0, 0
    
    for i in range(len(x)):
        n = n + (x[i] - avg_x) * (y[i] - avg_y)
        m = m + (avg_x - x[i]) ** 2
    
    return n / m

def get_b(avg_x, avg_y, a):
    return avg_y - (avg_x * a)

x = [1200, 1500, 1600, 1900, 2100] # 칼로리
y = [55, 61, 64, 72, 76] # 몸무게

avg_x = np.mean(x)
avg_y = np.mean(y)

a = get_a(x, y, avg_x, avg_y)
b = get_b(avg_x, avg_y, a)

res = [a * i + b for i in x]

plt.figure(figsize=(8,4))
plt.scatter(x, y, s=100)

plt.plot(x,res)
plt.plot(x,y)

plt.show()
```
