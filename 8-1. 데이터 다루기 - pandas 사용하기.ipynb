```python
import pandas as pd
```

```python
df = pd.read_csv('deeplearning/dataset/pima-indians-diabetes.csv',
                names = ['pregnant', 'plasma', 'pressure', 'thickness',
                         'insulin','BMI', 'pedigree','age', 'class'])
df
```

```python
df.info() # 데이터에 대한 자세한 정보
```

```python
# count : 샘플 수
# mean : 평균
# std : 표준 편차
# min, max : 최소값, 최대값
# 25%, 50%, 75% : 백분위로 25%, 50%, 75%에 해당하는 값

df.describe() # 데이터의 특징을 조금 더 자세히 알 수 있는 함수
```

```python
df[['BMI','class']] # 데이터의 일부 캄럼만 보고 싶을 때
```

```python
# groupby() 함수를 통해 '6'을 기준으로 하는 새로운 그룹을 만든다.
# as_index=False 속성을 통해 '6' 정보 옆에 새로운 인덱스를 만들어 준다.
# groupby()를 통해 나온 결과의 평균값을 구한다.

df[['BMI','class']].groupby(['BMI'], as_index=False).mean()
```

```python
# groupby()를 통해 나온 결과의 평균값을 정렬한다.
# 정렬 방식은 ascending=True를 통해 오름차순으로 지정한다.

df[['BMI','class']].groupby(['BMI'], as_index=False).mean().sort_values(by='BMI', ascending=True)
```
