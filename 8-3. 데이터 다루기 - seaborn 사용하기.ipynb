```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 각 정보간 어떤 상관관계가 있는지 분석을 도와주는 라이브러리
```

```python
df = pd.read_csv('deeplearning/dataset/pima-indians-diabetes.csv')
df
```

```python
plt.figure(figsize=(12,12)) # 만약 똑같은 2개의 결과가 나온다면 이거 지우면 된다.

# heatmap() 함수는 두 항목씩 짝을 지어 각각 어떤 패턴으로 변화하는지 관찰하는 함수
# 두 항목이 전혀 다른 패턴으로 변화하고 있으면 0,
# 서로 비슷한 패턴으로 변화하고 있으면 1에 가까운 값을 출력한다.
# heatmap() 함수를 통해 각 항목 같의 상관관계를 나타낼 수 있다.

# vmax : 색상의 밝기 조절
# cmap : 미리 정해진 matplotlib 색상의 설정값 가져오기

sns.heatmap(df.corr(), linewidths=0.1, vmax=0.5, cmap=plt.cm.gist_heat, linecolor='white', annot=True)
plt.show()
```

```python
# 결과(class)와 가장 상관관계가 높은 것은 plasma인데
# 책을 보면 플라스마의 칼럼은 148이라는 것을 확인할 수 있다.

# 여기서 class를 의미하는 것은 '1'
# FaceGrid()는 여러개의 서브 플롯을 만드는데 적절한 함수
grid = sns.FacetGrid(df, col='1')
grid.map(plt.hist, '148', bins=10) # bins : 얼마나 구체적으로 표현할 것인지
plt.show()
```
