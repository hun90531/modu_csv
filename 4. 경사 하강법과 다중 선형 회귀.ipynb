```python
# 경사 하강법

x = np.array([2, 4, 6, 8]) # 공부 시간
y = np.array([81, 93, 91, 97]) # 성적

a = 0
b = 0

lr = 0.03 # 학습률
epochs = 2001 # 반복 횟수

for i in range(epochs):
    y_pred = a * x + b
    error = y - y_pred # 오차
    
    # 평균 제곱 오차를 a, b로 미분
    # 구해진 기울기의 반대 방향으로 이동시켜야 하므로
    # - 부호를 붙여준다.
    a_diff = -(2 / len(x)) * sum(error * x)
    b_diff = -(2 / len(x)) * sum(error)
    
    
    # 경사 하감법 알고리즘은 기울기에 학습률을 곱해 다음 데이터를 갱신한다.
    a = a - lr * a_diff
    b = b - lr * b_diff
    
    # 기울기와 b가 2.3과 79에 점점 수렴하는 것을 확인할 수 있다.   
    if i % 100 == 0:
        print('epoch = %d\t,' % i, end='')
        print('a = %.2f, b = %.2f,\t' % (a, b), end='')
        print('a_diff = %.2f, b_diff = %.2f' % (a_diff, b_diff))

        
y_pred = a * x + b

plt.figure(figsize=(12, 5))
plt.scatter(x, y, s=80)
plt.plot(x, y_pred, 'r')
plt.show()
```

```python
# 다중 선형 회귀를 위한 데이터

x1 = np.array([2, 4, 6, 8]) # 공부 시간
x2 = np.array([0, 4, 2, 3]) # 과외 횟수
y = np.array([81, 93, 91, 97]) # 성적

ax = plt.axes(projection='3d')
ax.set_xlabel('study_hour')
ax.set_ylabel('private_class')
ax.set_zlabel('score')

ax.dist = 9 # 객체와 카메라 사이의 거리

ax.scatter(x1, x2, y)
plt.show()
```

```python
x1 = np.array([2, 4, 6, 8]) # 공부 시간
x2 = np.array([0, 4, 2, 3]) # 과외 횟수
y = np.array([81, 93, 91, 97]) # 성적

a1 = 0
a2 = 0
b = 0

lr = 0.02
epochs = 1401

for i in range(epochs):
    y_pred = a1 * x1 + a2 * x2 + b # 예측 값
    error = y - y_pred # 오차
    
    a1_diff = -(2 / len(x1)) * sum(error * x1)
    a2_diff = -(2 / len(x2)) * sum(error * x2)
    b_diff = -(2 / len(y)) * sum(error)
    
    a1 = a1 - lr * a1_diff
    a2 = a2 - lr * a2_diff
    b = b - lr * b_diff
    
    if i % 100 == 0:
        print('epoch = %d,\t' % i, end='')
        print('a1 = %.2f, a2 = %.2f, b = %.2f,\t' % (a1, a2, b), end='')
        print('a_diff = %.2f, b_diff = %.2f' %(a_diff, b_diff))    


y_pred = a1 * x1 + a2 * x2 + b

print()
print(f'y = {a1} * x1 * {a2} * x2 + {b}')

# 다중 선형 회귀를 위한 데이터


ax = plt.axes(projection='3d')
ax.set_xlabel('study_hour')
ax.set_ylabel('private_class')
ax.set_zlabel('score')

ax.dist = 9 # 객체와 카메라 사이의 거리

ax.scatter(x1, x2, y_pred)
plt.show()
```
